### Step 3: Changes Performed
     

```python
#Step 3.1: Defining 'Gender' Type Characteristics;

Rider_Gender = []

for value in df_Citibike["Gender"]:
    if value == 1:
        Rider_Gender.append("Male")
    elif value == 2:
        Rider_Gender.append("Female")
    else:
        Rider_Gender.append("Unknown")
```

```python
df_Citibike['Rider_Gender'] = Rider_Gender
```

```python
df_Citibike['Rider_Gender'].value_counts(dropna = False)
```

```python
#Checking for New Column; 

df_Citibike.head()
```

<b>New Column:</b> 'Rider_Gender' Added


```python
#Step 3.2: Deleting Unnecessary Columns; 

    #'Trip_ID', 'Gender' and 'Birth_Year' will not be needed on this analysis

df_Citibike_Clean = df_Citibike.drop(columns = ['Trip_ID', 'Birth_Year', 'Gender'])
```
