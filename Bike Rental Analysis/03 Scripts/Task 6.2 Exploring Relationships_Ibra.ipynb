```python
#Step 2.5: Add Labels, a Legend, and Change the Size of the Heatmap; 

f = plt.figure(figsize=(14, 14)) # figure size 
plt.matshow(df_Citibike_Final.corr(), fignum=f.number) # type of plot
plt.xticks(range(df_Citibike_Final.shape[1]), df_Citibike_Final.columns, fontsize=12, rotation = 90) # x axis labels
plt.yticks(range(df_Citibike_Final.shape[1]), df_Citibike_Final.columns, fontsize=12) # y axis labels
cb = plt.colorbar() # add a colour legend (called colorbar)
cb.ax.tick_params(labelsize=14) # add font size
plt.title('Correlation Matrix', fontsize=12) # add title
```

```python
#Step 2.6: Creating a Subplot with Matplotlib;

f,ax = plt.subplots(figsize=(16,16))

# Create the correlation heatmap in seaborn by applying a heatmap onto the correlation matrix and the subplots defined above.

corr = sns.heatmap(df_Citibike_Final.corr(), annot = True, ax = ax) 

    # The `annot` argument allows the plot to place the correlation coefficients onto the heatmap.
```


<b>Outcome:</b> Seems to be a moderate relationship between 'Start_Station_Latitude' and 'End_Station_Latitude' with a Correlation Co-Efficient of 0.59.


### Step 5: Categorical Plots


```python
#Step 5.1: Creating a Histogram of 'Start_Station_Latitude' Variable; 

sns.histplot(df_Citibike_Final['Start_Station_Latitude'], bins = 20, kde = True)
```

```python
#Step 5.1: Creating a Histogram of 'End_Station_Latitude' Variable; 

sns.histplot(df_Citibike_Final['End_Station_Latitude'], bins = 20, kde = True)
```

```python
#Step 5.2: Defining Categorical Variables: "Start_Station_Latitude" with Categories; High, Mid & Low; 

df_Citibike_Final.loc[df_Citibike_Final['Start_Station_Latitude'] < 40.70, 'Distance'] = 'Low_Latitude'
df_Citibike_Final.loc[(df_Citibike_Final['Start_Station_Latitude'] >= 40.70) & (df_Citibike_Final['Start_Station_Latitude'] < 40.74), 'Distance'] = 'Mid_Latitude'
df_Citibike_Final.loc[df_Citibike_Final['Start_Station_Latitude'] >= 40.74, 'Distance'] = 'High_Latitude'
```
